```python
from google.colab import drive
drive.mount('/content/drive')
import numpy as np
import pandas as pd
import math
from sklearn import preprocessing
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
```

```python
raw=pd.read_csv('/content/drive/Shareddrives/Hanjaya-Putra Lab/Members/Paul Jeong/PCAforDaniel/daniel.csv', index_col=0)
newdf=np.power(2, -raw)
x=newdf.fillna(0)
```

```python
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x.values.T)
norm_x=x_scaled.T



pca = PCA()
pca.fit_transform(norm_x.T)
pca.explained_variance_ratio_
```

```python
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x.values.T)
norm_x=x_scaled.T



pca = PCA(n_components=2)
newd=pca.fit_transform(norm_x.T)
```

```python
loadings = pca.components_
loadings_df = pd.DataFrame({'PC1':loadings[0], 'PC2':loadings[1]})
loadings_df.index=x.index
loadings_df.to_csv('/content/drive/Shareddrives/Hanjaya-Putra Lab/Members/Paul Jeong/PCAforDaniel/loadings_data.csv')
```
